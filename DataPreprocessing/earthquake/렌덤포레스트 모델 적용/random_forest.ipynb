```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import joblib # 모델 저장을 위해 joblib 라이브러리를 사용합니다.

print("--- 1. 라이브러리 임포트 ---")

# 1. 데이터 로드
file_path = 'second_final_data_proceed.csv'
df = pd.read_csv(file_path)
print(f"--- 2. '{file_path}' 데이터 로드 완료 ---")

# 2. 특징(X)과 타겟(y) 변수 분리
# 'tsunami'를 예측 대상으로 설정합니다.
X = df.drop('tsunami', axis=1)
y = df['tsunami']
print("--- 3. 특징(X)과 타겟(y) 분리 완료 ---")

# 3. RandomForest 분류기 모델 초기화
# random_state=42는 결과를 동일하게 재현하기 위한 설정입니다.
model = RandomForestClassifier(random_state=42, n_estimators=100)
print("--- 4. RandomForest 모델 초기화 완료 ---")

# 4. 모델 학습
# X(특징)와 y(타겟) 데이터를 사용하여 모델을 학습시킵니다.
model.fit(X, y)
print("--- 5. 모델 학습 완료 ---")

# 5. 학습된 모델을 파일로 저장
model_filename = 'random_forest_tsunami_model.joblib'
joblib.dump(model, model_filename)

print(f"\n[최종 완료] 모델이 '{model_filename}' 파일로 성공적으로 저장되었습니다.")
```
